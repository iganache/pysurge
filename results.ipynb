```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tsim = 1
numframes = 10
delt = tsim / numframes
```

```python
dem = np.genfromtxt("dem.txt")
```

```python
depth = np.genfromtxt("test5/Depth.txt")
fig, ax = plt.subplots(1,1)
fig.set_size_inches(8,6) 
ax.plot(dem, 'k')
ax.plot(dem+depth)
plt.show()
```

```python
deposit = np.genfromtxt("test5/Deposit_depth.txt")
fig, ax = plt.subplots(1,1)
fig.set_size_inches(8,6) 
ax.plot(deposit)
# ax.set_ylim(-100, 100)
plt.show()
```

```python
depmovie = np.genfromtxt("test5/DepthMovie.txt")
denmovie = np.genfromtxt("test5/DensityMovie.txt")
velmovie = np.genfromtxt("test5/VelocityXMovie.txt")
phimovie = np.genfromtxt("test5/PhiMovie.txt")
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(8,6) 
for i in range(depmovie.shape[0]):
    ax.plot(depmovie[i], label=delt*(i+1))
    
# ax.set_ylim(0,40)
# ax.set_xlim(2.5, 12.5)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(8,6) 
for i in range(denmovie.shape[0]):
    plt.plot(denmovie[i], label=delt*(i+1))
    
# ax.set_ylim(60, 90)
ax.legend()
ax.set_ylabel("density")
plt.show()
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(8,6) 
for i in range(velmovie.shape[0]):
    plt.plot(velmovie[i], label=delt*(i+1))
    
# ax.set_ylim(-5, 40)
ax.legend()
ax.set_ylabel("velocity")
plt.show()
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(8,6) 
for i in range(phimovie.shape[0]):
    plt.plot(phimovie[i], label=delt*(i+1))
    
# ax.set_ylim(-5, 40)
ax.legend()
ax.set_ylabel("solid concentration")
plt.show()
```
