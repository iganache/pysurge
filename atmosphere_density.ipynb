```python
import numpy as np
import matplotlib.pyplot as plt
```

### <b> settling velocity </b>

```python
cdrag = 1
rho_s = 3000
phi_s = 0.05
rho_g = 0.61
rho = (rho_s*phi_s) + (1-phi_s * rho_g)
d50 = 1e-3
```

```python
print(rho)
```

### <b> gravity term </b>

```python
h = 10   ## 10 m
rho_a = np.linspace(1, 65, 20)
g = 9.8
slope_rad = np.deg2rad(3)
gt = (rho-rho_a)/ rho_a * g * np.sin(np.arctan(slope_rad)) * h
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(10, 8)
ax.plot(rho_a, gt)
ax.scatter(rho_a[0], gt[0])
ax.scatter(rho_a[-1], gt[-1])
plt.show()
```
